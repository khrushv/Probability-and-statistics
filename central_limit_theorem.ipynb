```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
#В качестве непрерывного распределения выберем экспоненциальное распределение с параметром λ = 1
#Пусть случайная величина X ~ Exp(1)
exp_rv = sts.expon(loc=0, scale=1)
sample = exp_rv.rvs(1000)
```

```python
#Строим гистограмму выборки размера 1000 и теоретическую плотность вероятности 
x = np.linspace(0,7,100)
pdf = exp_rv.pdf(x)
plt.hist(sample, density = True, bins = 20, label = 'empirical pdf')
plt.plot(x, pdf, label = 'theoretical pdf')
plt.legend()
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
```

Мы знаем, что мат ожидание экспоненциального распределения равно $λ^{-1}$, дисперсия равна $λ^{-2}$,
так как λ = 1, то $E[X]$ = 1, σ = 1. По ЦПТ выборочное среднее приближается к нормальному с параметрами 
N($\mu= E[X]$ и $\sigma = \sigma_x/\sqrt(n)$)

```python
for n in [5, 10, 25, 50 , 250]:
    #Массив, содержащий посчитанные выборочные средние для тысячи выборок порядка n
    sample_means = np.empty(1000)
    for i in range(sample_mean.size):
        #Генерируем выбору нашего распределения и считаем выборочное среднее
        sample_means[i] = np.mean(exp_rv.rvs(n))
    plt.hist(sample_means, density = True, label = 'empirical pdf', bins = 10)
    #Создаем случайную величину с нормальным распределением, параметры задаем согласно тексту выше
    norm_rv = sts.norm(1, 1./np.sqrt(n))
    x = np.linspace(exp_rv.mean() - 3*exp_rv.std(), exp_rv.mean() + 3*exp_rv.std(), 1000)
    pdf_of_norm_rv = norm_rv.pdf(x)
    plt.plot(x, pdf_of_norm_rv, label = 'theoretical pdf')
    plt.ylabel('fraction of samples')
    plt.xlabel('sample mean')
    plt.legend()
    plt.title('n = %d' % n)
    plt.show()
```

Легко заметить, что с ростом n точность аппроксимации распределения выборочных средних нормальным распределением растет пропорционально корню из n. Так как стандартное отклонение на каждой итерации уменьшается в корень из n.
